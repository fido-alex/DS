1) Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#На вход данной функции поступает строка c образованием.
def education_university(edu):
    #Разбиваем строку с образованием по пробелам и присваиваем переменной edu_list
    edu_list = edu.split(' ')
    #Делаем срез edu_list, оставляем только первые 2 элемента.
    edu_list = edu_list[0:2]
    #Из списка формируем строку и возвращаем её
    edu_str = edu_list[0] + " " + edu_list[1]
    return edu_str

#Теперь применим эту функцию к столбцу c адресом.
hh_database['Образование'] = hh_database['Образование и ВУЗ'].apply(education_university)
#Посчитаем сколько соискателей имеют средний уровень образования
hh_database[hh_database['Образование'] == 'Среднее образование'].shape[0]
#Удалим признак 'Образование и ВУЗ'
hh_database = hh_database.drop('Образование и ВУЗ', axis=1)
```

* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#На вход данной функции поступает строка c полом и возрастом.
def gender(gender_age):
    #Разбиваем строку с полом и возрастом по пробелам и присваиваем переменной gender_list
    gender_list = gender_age.split(' ')
    #Смотрим на первый индекс в списке gender_list, в нём содержится пол соискателя.
    #Если в нём содержится 'Мужчина' заменяем на 'М', если 'Женщина' заменяем на 'Ж'.
    if gender_list[0] == 'Мужчина':
        gender_list[0] = 'М'
    elif gender_list[0] == 'Женщина':
        gender_list[0] = 'Ж'
    #Возвращаем пол соискателя
    return gender_list[0]

#Теперь применим эту функцию к столбцу Пол, возраст.
hh_database['Пол'] = hh_database['Пол, возраст'].apply(gender)

#На вход данной функции поступает строка c полом и возрастом.
def age(age):
    #Разбиваем строку с полом и возрастом по пробелам и присваиваем переменной age_list
    age_list = age.split(' ')
    #Смотрим на третий индекс в списке age_list, в нём находится возраст соискателя.
    #Возвращаем возраст соискателя
    return int(age_list[3])

#Теперь применим эту функцию к столбцу Пол, возраст.
hh_database['Возраст'] = hh_database['Пол, возраст'].apply(age)


female_resumes = hh_database[hh_database['Пол'] == 'Ж'].shape[0]
total_resume = hh_database['Пол'].shape[0]
print('Процент женских резюме:', round(female_resumes / (total_resume / 100) , 2))
print('Средний возраст соискателей:', round(hh_database['Возраст'].mean(), 1))
#Удалим признак 'Пол, возраст'
hh_database = hh_database.drop('Пол, возраст', axis=1)
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python

#На вход данной функции поступает строка c признаком Город, переезд, командировки.
def city(city_moving):
    #Разбиваем строку по пробелам и возвращаем первый элемент из списка, это и будет город.
    city_list = city_moving.split(' ')
    return city_list[0]

#Теперь применим эту функцию к столбцу Город, переезд, командировки.
city = hh_database['Город, переезд, командировки'].apply(city)
#Найдём в city все города из списка городов-милионников и заменим их строкой 'город-милионник'
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
city_million = city.apply(lambda x: 'город-миллионник' if x in million_cities else x)
#Найдём в city_million Москву, Санкт-Петербург, город-миллионник их оставим а все остальные города заменим на other.
hh_database['Город'] = city_million.apply(lambda x: x if x in ['Москва', 'Санкт-Петербург', 'город-миллионник'] else 'other')

#Сколько процентов соискателей живут в Санкт-Петербурге?
St_Petersburg = hh_database[hh_database['Город'] == 'Санкт-Петербург'].shape[0]
all_cities = hh_database['Город'].shape[0]
percent_St_Petersburg = St_Petersburg / (all_cities / 100)
print('В Петербурге живёт следующий процент соискателей:', round(percent_St_Petersburg))


#На вход данной функции поступает строка c признаком Город, переезд, командировки.
def willing_to_relocate(relocate):
    #Разбиваем строку по запятой.
    relocate_list = relocate.split(',')
    #Проходимся в цикле по всем элементам списка полученого на предыдущем шаге.
    for volume in relocate_list:
        #Убираем пробел в начале строки каждого элемента из списка.
        volume = volume.lstrip()
        #Ищем в начале строки volume указанную ниже фразу. 
        #В зависимости от того какая фраза найдена возвращаем True или False.
        if volume.startswith('не готова к переезду'):
            return False
        elif volume.startswith('не готов к переезду'):
            return False
        elif volume.startswith('готов к переезду'):
            return True
        elif volume.startswith('готова к переезду'):
            return True
        elif volume.startswith('хочу переехать'):
            return True 
    
#Теперь применим эту функцию к столбцу Пол, возраст.
hh_database['Готовность к переезду'] = hh_database['Город, переезд, командировки'].apply(willing_to_relocate)


#На вход данной функции поступает строка c признаком Город, переезд, командировки.
def willingness_to_travel(business_trips):
    #Разбиваем строку по запятой.
    business_trips_list = business_trips.split(',')
    #Проходимся в цикле по всем элементам списка полученого на предыдущем шаге.
    for volume in business_trips_list:
        #Убираем пробел в начале строки каждого элемента из списка.
        volume = volume.lstrip()
        #Ищем в начале строки volume указанную ниже фразу. 
        #В зависимости от того какая фраза найдена возвращаем True или False.
        if volume.startswith('не готов к командировкам'):
            return False
        elif volume.startswith('не готова к командировкам'):
            return False
        elif volume.startswith('не готов к редким командировкам'):
            return False
        elif volume.startswith('не готова к редким командировкам'):
            return False
        elif volume.startswith('готов к командировкам'):
            return True
        elif volume.startswith('готова к командировкам'):
            return True
        elif volume.startswith('готов к редким командировкам'):
            return True
        elif volume.startswith('готова к редким командировкам'):
            return True
    #Для всех остальных случаев возвращаем False
    return False
    
#Теперь применим эту функцию к столбцу Город, переезд, командировки.
hh_database['Готовность к командировкам'] = hh_database['Город, переезд, командировки'].apply(willingness_to_travel)

#Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам? Ответ округлите до целого.
travel = hh_database[(hh_database['Готовность к переезду'] == True) & (hh_database['Готовность к командировкам'] == True)].shape[0]
all_applicants = hh_database['Готовность к переезду'].shape[0]
percent_travel = travel / (all_applicants / 100)
print('Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам:', round(percent_travel))

#Удалим признак 'Город, переезд, командировки'
hh_database = hh_database.drop('Город, переезд, командировки', axis=1)
```

```python
#Передаём в функцию строку из столбца и аргумент по которому будет производиться поиск.
def employment(line, volume):
    #Разбиваем строку по запятым.
    line_list = line.split(',')
    #Проходимся по элементам списка созданного предыдущей строкой.
    for vol in line_list:
        #Убираем пробел в начале строки каждого элемента из списка.
        vol = vol.lstrip()
        #Осуществляем поиск и в случае удачи возвращаем True
        if vol.startswith(volume):
            return True
    return False

#Теперь применим эту функцию к столбцу Занятость.
hh_database['полная занятость'] = hh_database['Занятость'].apply(employment, volume='полная занятость')
hh_database['частичная занятость'] = hh_database['Занятость'].apply(employment, volume='частичная занятость')
hh_database['проектная работа'] = hh_database['Занятость'].apply(employment, volume='проектная работа')
hh_database['стажировка'] = hh_database['Занятость'].apply(employment, volume='стажировка')
hh_database['волонтерство'] = hh_database['Занятость'].apply(employment, volume='волонтерство')
#Теперь применим эту функцию к столбцу Занятость.
hh_database['полный день'] = hh_database['График'].apply(employment, volume='полный день')
hh_database['сменный график'] = hh_database['График'].apply(employment, volume='сменный график')
hh_database['гибкий график'] = hh_database['График'].apply(employment, volume='гибкий график')
hh_database['удаленная работа'] = hh_database['График'].apply(employment, volume='удаленная работа')
hh_database['вахтовый метод'] = hh_database['График'].apply(employment, volume='вахтовый метод')

project_volunteering = hh_database[(hh_database['проектная работа'] == True) & (hh_database['волонтерство'] == True)].shape[0]
print('Сколько людей ищут проектную работу и волонтёрство:', project_volunteering)
shiftMethod_flexibleSchedule= hh_database[(hh_database['вахтовый метод'] == True) & (hh_database['гибкий график'] == True)].shape[0]
print('Сколько людей хотят работать вахтовым методом и с гибким графиком:', shiftMethod_flexibleSchedule)

#Удалим признаки График, Занятость.
hh_database = hh_database.drop('График', axis=1)
hh_database = hh_database.drop('Занятость', axis=1)
```

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#Подгрузим таблицу с котировками валют и удалим ненужные столбцы
ExchangeRates = pd.read_csv('data/ExchangeRates.csv', sep=',')
ExchangeRates = ExchangeRates.drop('per', axis=1)
ExchangeRates = ExchangeRates.drop('time', axis=1)
ExchangeRates = ExchangeRates.drop('vol', axis=1)

#Приведём признак к формату datetime и получим дату обновления резюме
hh_database['Обновление резюме'] = pd.to_datetime(hh_database['Обновление резюме'], dayfirst=True)
hh_database['Обновление резюме'] = hh_database['Обновление резюме'].dt.date
hh_database['Обновление резюме'] = pd.to_datetime(hh_database['Обновление резюме'], dayfirst=True)
ExchangeRates['date'] = pd.to_datetime(ExchangeRates['date'], dayfirst=True)


#Функция, выделяет из признака ЗП цифру и наименование валюты
def zp(line, volume):
    #Разбиваем строку по запятым.
    line_list = line.split(' ')
    return line_list[volume]

#Теперь применим эту функцию.
hh_database['зарплата'] = hh_database['ЗП'].apply(zp, volume=0)
hh_database['валюта'] = hh_database['ЗП'].apply(zp, volume=1)

#Функция приводящая к формату ISO признак валюта
def currency(line):
    if line == 'бел.руб.':
        return 'BYN'
    elif line == 'грн.':
        return 'UAH'
    elif line == 'сум':
        return 'UZS'
    return line

#Теперь применим эту функцию к столбцу валюта.
hh_database['валюта'] = hh_database['валюта'].apply(currency)

#Заменим признаки: date на Обновление резюме, currency на валюта для ExchangeRates. Это нужно для корректного объединения по 2 столбцам таблиц.
ExchangeRates.rename(columns = {'date' : 'Обновление резюме', 'currency' : 'валюта' }, inplace = True)
#Объединим таблицы.
hh_database = hh_database.merge(
    ExchangeRates,
    how='left'
)

#Функция которая заполняет 1 пропуски в признаках close и proportion.
def close_r(close,currency):
    if currency == 'руб.':
        close = 1
        return close
    return close

#Применим функцию.
hh_database['close'] = hh_database.apply(lambda x: close_r(x['close'], x['валюта']), axis=1)
hh_database['proportion'] = hh_database.apply(lambda x: close_r(x['proportion'], x['валюта']), axis=1)

#Приведём к float признаки зарплата, close, proportion и посчитаем зарплату в рублях.
hh_database['зарплата'] = hh_database['зарплата'].astype('float64')
hh_database['close'] = hh_database['close'].astype('float64')
hh_database['proportion'] = hh_database['proportion'].astype('float64')
hh_database['ЗП (руб)'] = hh_database['зарплата'] * hh_database['close'] / hh_database['proportion']

print('Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях):', hh_database['ЗП (руб)'].median())

#Удалим признак ЗП
hh_database = hh_database.drop('ЗП', axis=1)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
#Строим гистрограмму распределения признака возраст в логарифмическом масштабе.
#На графике отмечаем среднее и границы интервала метода 3 сигм (справа 4 сигмы).
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
age = np.log(hh_database['Возраст'] + 1)
histplot = sns.histplot(age, bins=30, ax=ax)
histplot.axvline(age.mean(), color='k', lw=2)
histplot.axvline(age.mean()+ 4 * age.std(), color='k', ls='--', lw=2)
histplot.axvline(age.mean()- 3 * age.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение возраста');

#Функция для поиска выбросов с помощью метода z-отклонения.
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

#Применим функ функцию outliers_z_score к признаку Возраст. Выведем количество отклонений и результирующие 
#число записей. Выведем таблицу соискателей найденую по методу z-отклонений.
outliers, cleaned = outliers_z_score(hh_database, 'Возраст', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
display(outliers)

#Удалим выбросы найденные с помощью метода z-отклонений.
hh_database = hh_database.drop(outliers.index)
```

Распределение имеет правосторонюю асимметрию. Мы посчитали таблицу выбросов по методу z-отклонений. В неё попали соискатели с возрастом 15 и 100 лет. Дествительно законодательство РФ разрешает работать с 16 лет и в 100 лет никто не будет искать работу.
